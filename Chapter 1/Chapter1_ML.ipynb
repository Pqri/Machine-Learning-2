```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Dataset sederhana dari buku (GDP per capita vs Life Satisfaction)
data = {
    "Country": ["Hungary", "Korea", "France", "Australia", "United States"],
    "GDP per capita": [12240, 27195, 37675, 50962, 55805],
    "Life satisfaction": [4.9, 5.8, 6.5, 7.3, 7.2]
}

df = pd.DataFrame(data)
```

```python
# Menampilkan data
print("Dataset:")
print(df)
```

```python
# Visualisasi data
plt.scatter(df["GDP per capita"], df["Life satisfaction"], color='blue')
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life Satisfaction Score")
plt.title("GDP per Capita vs Life Satisfaction")
plt.grid(True)
plt.show()
```

```python
# Persiapan data untuk model
X = df[["GDP per capita"]]  # fitur (independen variable)
y = df[["Life satisfaction"]]  # target (dependent variable)

# Membuat dan melatih model Linear Regression
model = LinearRegression()
model.fit(X, y)
```

```python
# Prediksi untuk negara Cyprus dengan GDP per capita = 22587 USD
X_new = [[22587]]  # input baru
prediction = model.predict(X_new)
print(f"\nPrediksi Life Satisfaction untuk Cyprus (GDP per Capita = 22,587 USD): {prediction[0][0]:.2f}")
```

```python
# Visualisasi hasil model
plt.scatter(df["GDP per capita"], df["Life satisfaction"], color='blue', label='Data Aktual')
plt.plot(df["GDP per capita"], model.predict(X), color='red', linewidth=2, label='Model Linear')
plt.scatter(X_new, prediction, color='green', marker='x', s=100, label='Prediksi Cyprus')
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life Satisfaction Score")
plt.title("Linear Regression Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

# Chapter 1 Summary

**Objective**: Memprediksi skor kepuasan hidup (Life Satisfaction) menggunakan GDP per capita dengan model Linear Regression.

## Machine Learning Concepts:
1. **Supervised Learning** digunakan karena kita melatih model dengan label target.
2. **Regression Task**: Target berupa nilai kontinu.
3. **Linear Model**: Life_satisfaction = θ0 + θ1 × GDP_per_capita
4. **Training**: Mencari parameter optimal (θ0 dan θ1) yang meminimalkan error.
5. **Evaluation**: Prediksi diuji untuk GDP negara Cyprus.
6. **Overfitting vs Underfitting**:
   - Model sederhana → risiko underfitting (tidak cukup menangkap pola data).
   - Model kompleks → risiko overfitting (terlalu cocok dengan data training).

## Output:
- Prediksi Life Satisfaction untuk Cyprus (GDP per Capita = 22,587 USD) ~ 5.96

## 16. What is a test set, and why would you want to use it?
A test set is a dataset used only for evaluating the final performance of a trained model to estimate its ability to generalize to new data.

---

## 17. What is the purpose of a validation set?
A validation set is used to tune hyperparameters and select the best model before testing on the test set.

---

## 18. What is the train-dev set, when do you need it, and how do you use it?
- The train-dev set is a subset of the training data used when the test set is too small to split into both validation and test sets.
- It's used to perform quick evaluations during development without touching the final test set.

---

## 19. What can go wrong if you tune hyperparameters using the test set?
The model may overfit to the test set, leading to overly optimistic performance estimates that do not generalize to truly unseen data.
